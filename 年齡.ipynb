```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sys, cv2
imagePath=(r"C:\Python\audi(2).jpg")
image=cv2.imread(imagePath)
```

```python
print(image.shape)
```

```python
plt.imshow(image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
face_Cascade=cv2.CascadeClassifier(r"C:\Users\joann\anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")

gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces=face_Cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30,30),
    flags = cv2.CASCADE_SCALE_IMAGE)

print (faces)
```

```python
import random
font=cv2.FONT_HERSHEY_SIMPLEX

for (x, y, w, h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (14, 201, 255), 2)
    cv2.putText(image,str(random.randrange(20,25)), (x+(w//2)-18,y-10),font,3,(28,230,285),3)
    cv2.rectangle(image, (image.shape[1]-280, image.shape[0]-40),
                  (image.shape[1], image.shape[0]),(0,255,255),-1)
    cv2.putText(image, "Finding "+ str(len(faces)) + " face ",
                (image.shape[1]-260, image.shape[0]-10),
                cv2.FONT_HERSHEY_TRIPLEX, 1, (255,0,0), 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
from PIL import Image
```

```python
image_path = r"C:\Python\audi(2).jpg"
```

```python
num = 1
for (x,y,w,h) in faces:
    filename = "face" + str(num) +".jpg"
    image= Image.open(image_path)
    imageCrop=image.crop((x,y,x+w,y+h))
    imageResize=imageCrop.resize((150,150),Image.Resampling.LANCZOS)
    imageResize.save(filename)
    num +=1
```
